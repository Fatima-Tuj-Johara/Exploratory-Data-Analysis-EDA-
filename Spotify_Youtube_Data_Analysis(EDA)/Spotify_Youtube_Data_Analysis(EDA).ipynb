```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data = pd.read_csv("Spotify_Youtube Dataset.csv")
```

```python
data
```

```python
data.columns
```

```python
data.info()
```

```python
data.drop(columns = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace = True)
```

```python
data.head(2)
```

```python
# checking missing values

data.isna().sum()
```

```python
# filling the missing values with 0 in Likes & Commnets column

data['Likes'] =  data['Likes'].fillna(0)
data['Comments'] = data['Comments'].fillna(0)
```

```python
data.isnull().sum()
```

```python
text_columns = ['Title', 'Channel', 'Description']
data[text_columns] = data[text_columns].fillna('Unknown')
```

```python
bool_columns = ['Licensed', 'official_video']
data[bool_columns] = data[bool_columns].fillna(False).astype(bool)
```

```python
data.dropna(subset=['Stream', 'Views'], how='any', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

####  Top 10 Artists - with the Highest Views on YouTube.

```python
data.head(2)
```

```python
Artist_grouped=  data.groupby('Artist')['Views'].sum()
```

```python
Artist_sorted =  Artist_grouped.sort_values(ascending = False)
```

```python
Artist_sorted.head(10)
```

```python
top_10_artists = Artist_sorted.head(10)

plt.figure(figsize=(12, 6))

sns.barplot(x=top_10_artists.values, y=top_10_artists.index, palette="viridis")

plt.title('Top 10 Artists with the Highest Views on YouTube', fontsize=16)
plt.xlabel('Total Views (in billions)', fontsize=12)
plt.ylabel('Artist', fontsize=12)

plt.show()
plt.tight_layout()
```

#### Top 10 Tracks - with the Highest Streams on Spotify.

```python
data.head(1)
```

```python
x = data[['Track', 'Stream']]

x
```

```python
most_stream_track =  x.sort_values(by = ['Stream'], ascending=False)
```

```python
most_stream_track.head(10)
```

```python
top_10_tracks = most_stream_track.head(10)
plt.figure(figsize=(12, 6))

sns.barplot(
    x=top_10_tracks['Stream'],
    y=top_10_tracks['Track'],
    hue=top_10_tracks['Track'],
    palette="magma",
    legend=False
)

plt.title('Top 10 Most Streamed Tracks on Spotify', fontsize=16)
plt.xlabel('Total Streams (in billions)', fontsize=12)
plt.ylabel('Track Title', fontsize=12)

plt.show()
plt.tight_layout()
```

#### The most common Album Types on Spotify. How many tracks belong to each album type?

```python
data.Album_type.unique()
```

```python
album_type
```

####  How do the Average Views, Likes, and Comments are compared between different Album Types?

```python
df = data.groupby('Album_type')[['Likes', 'Views', 'Comments']].mean()

df
```

```python
df = df.reset_index()

df
```

```python
df_melted  = pd.melt( df, id_vars = 'Album_type', var_name = "Attribute", value_name = 'Total' )

df_melted
```

```python
plt.figure(figsize = (9,4))

sns.barplot( x = 'Album_type', y = 'Total', hue = 'Attribute', data = df_melted );

plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(x='Album_type', y='Total', hue='Attribute', data=df_melted, palette='Set2')
plt.yscale('log')

plt.title('Comparison of Attributes by Album Type (Log Scale)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Album Type', fontsize=12)
plt.ylabel('Total Count (Log Scale)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.legend(title='Attribute', frameon=True, facecolor='white', edgecolor='none')

plt.tight_layout()
plt.show()
```
